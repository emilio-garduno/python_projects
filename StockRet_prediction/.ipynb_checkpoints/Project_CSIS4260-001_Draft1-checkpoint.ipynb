#### Clean the headlines dataset

Our Date data is in text (string) format. We want to change it to a datetime format so that it is easier to run our analysis along with our stock price in the following steps. Before we can modify the date using code, we need to briefly look through the dataset to have a sense of the format of the data. For instance...

**1.** Some rows might not have a year. For example, the website SeekingAlpha doesn’t include the year if the article is published in the current year of request. (e.g. Wed, Dec. 4 **[NoYear]**)

**2.** We are not interested in the day (*Monday, Tuesday... and so forth*), we just want the date (DD-MM-YYYY). Thus, we need to extract the Date and add in the Year (e.g. **Wed,** Dec. 4 )

**3.** All months except May have a Period symbol (**full point**) after it, e.g. “Jan.”, “Feb.” etc. The full point exists to indicate the spelling of the month is truncated. The month of May doesn’t need this. Thus, in our Regex code, we do not need to include a Period symbol.

**4.** We might have two variations from the previous bullet point. One for the dates with year (e.g. May 17, 2018) and one for dates without year (e.g. May 13, **[No Year]**)

**5.** First three rows don't contain a day or date. They say "Today". We change these dates via hard coding since it is inefficient to create a systematic code when it will only be used once

```python
headlines['Date'].iloc[0:3] = 'Mar. 1'
headlines.head()
```

#### Convert cleaned date to determine format

Once we identified all the dates in either “MMM. DD, YYYY” or “May DD, YYYY” format, it is time to convert these to datetime format. We use `datetime.strptime()` method to convert date to time.

Here we set a loop to clean the entire "Date" column going through every row and look for any of the 4 date string variations according to the previous description.

```python
newDateList = [] # create a list to store the cleaned dates

"""
Variations:

'r'
\w{3}\.\s\d{1,2}    
\w{3}\.\s\d{1,2}\,\s\d{4} 
    
    The 'r' at the start of the pattern string designates a python "raw" string which passes through backslashes 
    without change which is very handy for regular expressions
    
    \w metacharacter matches word characters. 
    
        \w{3} looks for 3 letters
        \. looks for a period symbol
        \s means “a string of whitespace characters, a space”
        \d{1,2} looks for 1 or 2 digits
        \d{4} to grab the year
            
| pipe symbol means 'or'
    
May\s\d{1,2}    
May\s\d{1,2}\,\s\d{4}
        
        May 
        \s means “a string of whitespace characters, or space”
        \d{1,2} looks for 1 or 2 digits
        \d{4} to grab the year
                                            
"""

for dateOfArticles in headlines['Date']: # loop every row in the "Date" column
    match = re.search(r'\w{3}\.\s\d{1,2}\,\s\d{4}|May\s\d{1,2}\,\s\d{4}|\w{3}\.\s\d{1,2}|May\s\d{1,2}', 
                      dateOfArticles)
    
#Once we found the variation, we check if it contains the year.
#If yes, don’t add a year to the string. If no, add the appropriate year to the end of the string.

    if re.search(r'\w{3}\.\s\d{1,2}\,\s\d{4}|\w{3}\s\d{1,2}\,\s\d{4}',match[0]):
        fulldate = match[0] # don't append year to string
    else:
        fulldate = match[0] + ", 2023" # append year to string

#Next, we convert the “Date” data from string to datetime format.
#Our dates have 2 possible formats now, one with a period symbol and one without. We will check for both.

    for fmt in ('%b. %d, %Y', '%b %d, %Y'):
        try:
            newDate = datetime.strptime(fulldate, fmt).date()
            break # if format is correct, don't test any other formats
        except ValueError:
            pass

#Once done, add the new date data to a list
        
    newDateList.append(newDate) # add new date to the list

"""
Check if the number of rows of the list match with original dataframe.
If yes, add the list as new column to our original dataframe.
"""

if(len(newDateList) != headlines.shape[0]):
    print("Error: Rows don't match")
else:
    headlines['New Date'] = newDateList # add the list to our original dataframe

headlines
```

#### Run sentiment analysis and create a score index

The next part is to send our headlines into a sentiment analyser to churn out a score. We use a loop to pass every headline into our analyser.

```python
results = []

for headline in headlines['Title']:
    pol_score = SIA().polarity_scores(headline) # run analysis. A sentiment score is assigned to each headline.
    pol_score['headline'] = headline # add headlines for viewing
    results.append(pol_score)
```

We add the Vader scores to the headlines dataset

```python
headlines[['Neg','Neu','Pos','Compound']] = pd.DataFrame(results)[['neg','neu','pos','compound']]
headlines
```

The positive, negative, and neutral scores are ratios for the proportions of text that fall in each category and should sum to
**1**. The compound score is the sum of positive, negative & neutral scores which is then normalized between -1(most extreme negative) and +1 (most extreme positive). The more Compound score closer to +1, the higher the positivity of the text. 

In general and mostly, the compound score is used as a threshold value for the analysis of the text data. We can even leverage the flexibility of changing the compound score (threshold value) and label the text data. This merely depends on business requirements and domain knowledge. 

Below is the standard scoring metric followed by most of the analyzers.

    -Positive sentiment: compound score >= 0.05
    -Neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
    -Negative sentiment: compound score <= -0.05
    
**Being said that, it is convenient to work with the values from the latter column (Compound) for sentiment analysis.**

```python
hlines_scores = headlines.drop(['Neg', 'Neu','Pos'], axis=1) 
hlines_scores = hlines_scores.rename(columns={'Compound':'Score'}, inplace = False) #Rename the column
hlines_scores
```

#### Get the Polarity and Subjectivity

```python
#Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

#Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
hlines_scores['Subjectivity'] = hlines_scores['Title'].apply(getSubjectivity)
hlines_scores['Polarity'] = hlines_scores['Title'].apply(getPolarity)
hlines_scores.head()
```

All news items corresponding to a day were concatenated as one text input for the Sentiment Analyzer returning the applicable Sentiment cloud for the date. The output will be the **New Date (as your index)** and the daily scores.

```python
hlines_scores = hlines_scores.groupby(['New Date']).sum() # creates a daily score by summing the scores of the individual articles in each day
hlines_scores
```

Now that we have our prices, we need to calculate our returns.
To calculate daily returns, we divide today’s prices by yesterday’s prices.

**We must keep the previously cleaned data: `dfEodPrice` for further use** 

Log Returns have this nice feature that they are much more Stationary compared to Prices, which means that they are much better candidates for building a time series model using LSTM. Using returns is also closely related to how time series models actually work.   

Why is Price misleading?
Suppose that my price predictions are actually very accurate with a small MSE. Even with this small error the predictions are always in the wrong direction. For example, suppose that today's price is $100 and it goes down to $99.99 The prediction is that it will go up to $100.01 That's a very small error no matter how we measure it. However, if the predictions are always in the wrong direction, the investor will always be losing money.
In fact, it's possible to always lose money even if the price prediction errors are small 

```python
dfReturns = dfEodPrice.copy() 
dfReturns['Returns'] = dfReturns['Adj Close']/dfReturns['Adj Close'].shift(1) - 1 # calculate daily returns
dfReturns
```

#### Check relationship between lagged score against returns

*Steps*:

    1. Lagged the sentiment score
    2. Match the daily returns with the lagged sentiment score
    3. Clean the data (again)
    4. Design the test
    5. Test the predictive value

#### 1. Lagged the sentiment score

*Lag is the number of days before current day that sentiment score had causal effect on stock market return.*
This code shifts all the data (scores) down by one row. E.g. on 2019-01-07, the lagged score is 0.4404. When we run a regression of 0.4404 against the TSLA’s 2019-01-07 returns, we are in fact checking the 2019-01-03’s score against 2019-01-07’s returns.

    --> Older date’s score vs future returns. <--

```python
hlines_scores1 = hlines_scores.shift(1)
hlines_scores1
```

#### 2. Match the daily returns with the lagged sentiment score

```python
print('No. of rows from Scores dataframe:', hlines_scores1.shape[0])  #Sentiment scores
print('No. of rows from Returns dataframe:', dfReturns.shape[0]) #Returns 
```

*The number of rows of our Scores dataframe is not the same as the number of rows of our Returns dataframe*. This happens as there are some trading days where there isn’t any news. Thus, we need to match the daily returns against the corresponding sentiment scores before we can run the regression. The following code will create a new dateframe that uses TSLA returns as reference and pull the appropriate lagged sentiment score for it.

```python
dfReturnsScore = pd.merge(dfReturns[['Returns']], hlines_scores1, 
                       left_index=True, right_index=True, how='left')
dfReturnsScore
```

#### 3. Clean the data one more time

On days where there is no news, there are no sentiment scores. The score column will show a NaN (not-a-number) when there are no scores. Because NaN is the equivalent of having a score of 0, we replace all NaNs with 0.

```python
dfReturnsScore.fillna(0, inplace=True) 
# replace NaN with 0 permanently
dfReturnsScore
```

#### 4. Design the test
Arbitrary figure: We shall assume that a score of between -0.5 and 0.5 is insignificant for the sake of simplicity (If the compound score is above 0.05, the sentiment is overall positive. The higher the score, the better the sentiment. In opposite, if the compound score is lower than -0.05, the sentiment is negative. If the score is between -0.05 to 0.05, the sentiment is neutral). 

*However, this test doesn’t test if the score has any longer term effects as we are only comparing today’s score against tomorrow’s stock returns*.

The code below removes all data where the sentiment score is between -0.5 and 0.5.

```python
dfReturnsScore2 = dfReturnsScore[(dfReturnsScore['Score'] > 0.05) | 
                                 (dfReturnsScore['Score'] < -0.05)]

dfReturnsScore2
```

#### 5. Test the predictive value
We need to test if there is a positive relationship between the lagged sentiment scores and the daily returns. One way to do this is by checking for correlation

```python
dfReturnsScore2.plot(x="Score", y="Returns", style="o")
```

On the x-axis, we have our 1-day lagged sentiment score. On the y-axis we have our daily TSLA returns.
We are expecting an upward sloping shape. An upward sloping shape indicates that when 'Score' goes up, the daily 'Returns' go up, and vice versa.

```python
pearsonr(dfReturnsScore2['Returns'],dfReturnsScore2['Score'])
```

From the previous output, the first value corresponds to the correlation coefficient whereas the second one corresponds to the p-value. On the one hand, as the correlation coefficient (0.13) is significantly below 0.5, it can be concluded that the relationship between the sentiment (Compound Score) and the market returns at a given day *t* is relatively weak. That is, article headlines alone do not have any predictive value for tomorrow’s stock returns. 

- We are better off using a shorter time delay such as a 2 second lag. But note that data of such **low timeframes are expensive and might not be accurate**

On the other hand, the sentiment captured from headlines can be potentially used as a signal to predict the closing price if the result is statistically significant (p-value < 0.05) which is our case. Therefore, the current sentiment analysis architecture may reveal signals that help predicting market returns the next day.

### Define some functions to scale and split the data, as well as a function for model training

- **Data Pre-processing** is an important step in LSTM. Scaling of data is a process which is advisable with most models, thus LSTM also requires processing in the form of scaling.
- **WE ARE NOT USING "Min-Max" scaling** based on the idea that nobody may know what is the max of a certain price, especially in the stock market. Also, the dataset will never reflect the true lows or true highs. Prices are always increasing, there is no maximum or minimum value. Price has no upper bound. In contrast, **Robust Scaler uses the meadian and InterQuartile Range** rather than the min-max, which makes it resistant to outliers.

```python
"""
We use a function to scale the data
"""
rs = RobustScaler()
def scaler_function (x):
    scaled_df = rs.fit_transform(x) 
    return scaled_df
```

#### Split the data into Training and Testing Sets. 

```python
def split_function(dataset):
    split_ratio = 0.8
    split_index = int(split_ratio * len(dataset))

    train = dataset[:split_index]
    test = dataset[split_index:]
    
    return train, test
```

- **Build the Model using LSTM and add different layers**

We build the model cointaing 4 layers of LSTM network, all followed by a **Dropout layer**.  To reduce overfitting, the dropout layer just randomly takes a portion of the possible network connections. This value is the percentage of the considered network connections per epoch/batch. On the bottom, we also have a final **Dense layer**. 

```python
def create_model(inputshape):
    model = Sequential()

    #First Layer
    model.add(LSTM(units = 100, return_sequences=True, input_shape = inputshape)) #Layer of 100 units
    model.add(Dropout(0.2))

    #Second Layer
    model.add(LSTM(units = 100, return_sequences=True))
    model.add(Dropout(0.2))

    #Third Layer
    model.add(LSTM(units = 100, return_sequences=True))
    model.add(Dropout(0.2))

    #Fourth Layer
    model.add(LSTM(units = 100, return_sequences=False))
    model.add(Dropout(0.2))

    model.add(Dense(units=1)) #A dense layer with one output unit
    
    return model
```

```python
dfEodPrice_sc = scaler_function(dfEodPrice)
```

```python
train_df, test_df = split_function(dfEodPrice_sc)
```

- Since LSTM works on sequences using them as the base for prediction of single value, a matrix must be created from the train data set available. As a result of the entire model building exercise, a sliding historical window of 60 days gave the best results among the range covered. Thus, we prepare input sequences, with **60 timesteps** of X_train data alongside the respective y_train labels.
- Thereafter, we reshape the data by adding another access for the batch size, since the input for an LSTM network is a 3D tensor. It must have a sequence length, the timesteps and the batch size.

```python
"""
Create X_train and y_train data structures
"""

X_train = []
y_train = []
X_test = []

length_train = len(train_df)
length_test = len(test_df)

timestep = 60

for i in range(timestep, length_train): 
    X_train.append(train_df[i-timestep:i,0])
    y_train.append(train_df[i,0])

for i in range(timestep, length_test):
    X_test.append(test_df[i-timestep:i,0])

"""
Convert the X_train, y_train, X_test to numpy arrays
"""
   
X_train = np.array(X_train)
y_train = np.array(y_train)

print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)

X_test = np.array(X_test)
print('X_test shape:', X_test.shape)
print()

"""
Reshape the data
"""

X_train = np.reshape(X_train,(X_train.shape[0], X_train.shape[1], 1)) #adding the batch_size axis
print('X_train after reshape:', X_train.shape)

X_test = np.reshape(X_test,(X_test.shape[0], X_test.shape[1],1))
print('X_test after reshape:', X_test.shape)
```

The train data fed into the LSTM consists of a multi-dimensional array consisting of various instances of dependent variable and the corresponding linked independent variables, which in our case is an array consisting historical close prices.
Compile using an adam optimizer and the mean squared error as a loss function.

- **Fit the Model**

Train the data using 20 epochs (iterations) and batch_size of 32 using the `model.fit` function by parsing X_train and y_train data.


```python
input_shape = (X_train.shape[1],1)
model = create_model(input_shape) #We use the model function and the corresponding parameter input_shape 
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
model.fit(X_train, y_train, epochs=20, batch_size=32, verbose=2)

"""
'loss' refers to the loss value over the training data after each epoch. 
This is what the optimization process is trying to minimize with the training so, the lower, the better.
"""
```

- **Predict the model output using the `model.predict` function by passing the X_test data**
- Set the proper format of the results using `inverse_transform` function to plot the values with the actual and predicted data

```python
y_pred = model.predict(X_test)
predicted_prices = rs.inverse_transform(y_pred)
print('predicted_prices data shape:', predicted_prices.shape)

test_prices = rs.inverse_transform(test_df[60:])
print('test data shape:', test_prices.shape)
```

- Add the results to the corresponding list for further analysis. Get the R2, MSE, and MAE, as the quantifying parameters to evaluate the success of the model being tested 

```python
r2.append(r2_score(test_prices, predicted_prices)) #add results to the list
mse.append(mean_squared_error(test_prices, predicted_prices))
mae.append(mean_absolute_error(test_prices, predicted_prices))
model_name.append('LSTM - Prices Only')
print('Results added to the lists')
```

We don't even need to predict the amount just whether or not it will go up or down. 

STOCK MOVEMENTS ARE UNPREDICTABLE
Only historical stock prices will not help to predict the future. LSTM model leaves out all the historical data. 

Stationarity and Extrapolation
NOT GOOD IDEA - prices as input and output
Stationarity means the distribution of the points in your time series does not change over time. e.g. if the means or variance of the prices changes over time means non-stationary.
If the data is not stationary then the relationship between the variables could change with t(Time) which may make the model obsolete for the time period we actually want to use it. 
Extrapolation
Consider the stock from January 2021 - July 2022
Training (January 2021 - January 2022): $250 - $400
Testing (February 2022 - July 2022): $400 - $450
The reason this is an issue is because you're asking your model to make predictions when your inputs are OUTSIDE of the range of values it was trained on. Models like SVM, Random Forest, or even Neural Networks, fail to extrapolate even for simple patterns. Furthermore, they all extrapolate in different ways

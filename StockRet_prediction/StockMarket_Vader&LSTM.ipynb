## 3. Text data pre-processing 

### 3.1 Clean the headlines dataset
Our Date data is in text (string) format. We want to change it to a datetime format so that it is easier to run our analysis along with our stock price in the following steps. Before we can modify the date using code, we need to briefly look through the dataset to have a sense of the format of the data. 

**1.** Some rows might not have a year. For example, the website SeekingAlpha doesn’t include the year if the article is published in the current year of request. (e.g. Wed, Dec. 4 **[NoYear]**)

**2.** We are not interested in the day (*Monday, Tuesday... and so forth*), we just want the date (DD-MM-YYYY). Thus, we need to extract the Date and add in the Year (e.g. **Wed,** Dec. 4 )

**3.** All months except May have a Period symbol (**full point**) after it, e.g. “Jan.”, “Feb.” etc. The full point exists to indicate the spelling of the month is truncated. The month of May doesn’t need this. Thus, in our Regex code, we do not need to include a Period symbol.

**4.** We might have two variations from the previous bullet point. One for the dates with year (e.g. May 17, 2018) and one for dates without year (e.g. May 13, **[No Year]**)

**5.** First three rows don't contain a day or date. They say "Today". We change these dates via hard coding since it is inefficient to create a systematic code when it will only be used once

```python
headlines['Date'].iloc[0:3] = 'Mar. 1'
headlines.head()
```

### 3.2 Convert cleaned date to determine format

Once we identified all the dates in either “MMM. DD, YYYY” or “May DD, YYYY” format, it is time to convert these to datetime format. We use `datetime.strptime()` method to convert date to time.

Here we set a loop to clean the entire "Date" column going through every row and look for any of the 4 date string variations according to the previous description.

```python
newDateList = [] # create a list to store the cleaned dates

"""
Variations:

'r'
\w{3}\.\s\d{1,2}    
\w{3}\.\s\d{1,2}\,\s\d{4} 
    
    The 'r' at the start of the pattern string designates a python "raw" string which passes through backslashes 
    without change which is very handy for regular expressions
    
    \w metacharacter matches word characters. 
    
        \w{3} looks for 3 letters
        \. looks for a period symbol
        \s means “a string of whitespace characters, a space”
        \d{1,2} looks for 1 or 2 digits
        \d{4} to grab the year
            
| pipe symbol means 'or'
    
May\s\d{1,2}    
May\s\d{1,2}\,\s\d{4}
        
        May 
        \s means “a string of whitespace characters, or space”
        \d{1,2} looks for 1 or 2 digits
        \d{4} to grab the year
                                            
"""

for dateOfArticles in headlines['Date']: # loop every row in the "Date" column
    match = re.search(r'\w{3}\.\s\d{1,2}\,\s\d{4}|May\s\d{1,2}\,\s\d{4}|\w{3}\.\s\d{1,2}|May\s\d{1,2}', 
                      dateOfArticles)
    
#Once we found the variation, we check if it contains the year.
#If yes, don’t add a year to the string. If no, add the appropriate year to the end of the string.

    if re.search(r'\w{3}\.\s\d{1,2}\,\s\d{4}|\w{3}\s\d{1,2}\,\s\d{4}',match[0]):
        fulldate = match[0] # don't append year to string
    else:
        fulldate = match[0] + ", 2023" # append year to string

#Next, we convert the “Date” data from string to datetime format.
#Our dates have 2 possible formats now, one with a period symbol and one without. We will check for both.

    for fmt in ('%b. %d, %Y', '%b %d, %Y'):
        try:
            newDate = datetime.strptime(fulldate, fmt).date()
            break # if format is correct, don't test any other formats
        except ValueError:
            pass

#Once done, add the new date data to a list
        
    newDateList.append(newDate) # add new date to the list

"""
Check if the number of rows of the list match with original dataframe.
If yes, add the list as new column to our original dataframe.
"""

if(len(newDateList) != headlines.shape[0]):
    print("Error: Rows don't match")
else:
    headlines['New Date'] = newDateList # add the list to our original dataframe

headlines
```

## 4. Run the Sentiment Analysis 

### 4.1 Get the Scores from the Vader Sentiment Intensity Analyser
The `Sentiment Intensity Analyzer`**(SIA)** library is used to generate scores which gives 4 values corresponding to the input text. It measures the level of Positive, Negative, Neutral and Compound sentiment associated with the input text. Therefore, we send our headlines into the sentiment analyser to churn out a score. We use a loop to pass every headline into our analyser.

```python
results = []

for headline in headlines['Title']:
    pol_score = SIA().polarity_scores(headline) # run analysis. A sentiment score is assigned to each headline.
    pol_score['headline'] = headline # add headlines for viewing
    results.append(pol_score)
    
"""
We add the Vader Scores to the headlines dataset
"""
headlines[['Neg','Neu','Pos','Compound']] = pd.DataFrame(results)[['neg','neu','pos','compound']]
headlines
```

The positive, negative, and neutral scores are ratios for the proportions of text that fall in each category and should sum to
**1. The compound score is the sum of positive, negative & neutral scores which is then normalized between -1(most extreme negative) and +1 (most extreme positive). The more Compound score closer to +1, the higher the positivity of the text.**

Being said that, it is convenient to work with the values from the latter column (Compound) for sentiment analysis.

```python
hlines_scores = headlines.drop(['Neg', 'Neu','Pos'], axis=1) 
hlines_scores = hlines_scores.rename(columns={'Compound':'Score'}, inplace = False) #Rename the column
hlines_scores.head()
```

### 4.2 Get the Polarity and Subjectivity from TextBlob

The polarity index is where every statement is quantified with a positive or negative value. From the polarity sign it’s inferred if the statement is positive, negative and neutral. The subjectivity index is to depict how relatable the sentence is to the company

```python
#Create a function to get the subjectivity
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity

#Create a function to get the polarity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity
```

```python
hlines_scores['Subjectivity'] = hlines_scores['Title'].apply(getSubjectivity)
hlines_scores['Polarity'] = hlines_scores['Title'].apply(getPolarity)
hlines_scores.head()
```

### 4.3 Create a daily score 

Since there is more than one news items for a single day, all news items for a single day are concatenated to arrive at the combined news data for the day. The output will be the **New Date** column **(as index)** and the scores of the articles in each day.

```python
hlines_scores = hlines_scores.groupby(['New Date']).sum() # creates a daily score by summing the scores of the individual articles in each day
hlines_scores.head()
```

### 7.1 Data pre-processing for Model Building
#### 7.1.1 Function to Scale the data
- Standardization of data is a process which is advisable with most models including LSTM. Thus, it may require pre-processing in the form of scaling

- **WE ARE NOT USING "Min-Max" scaling** based on the assumption that nobody may know what is the max value of a certain price, especially in the stock market. In addition, the dataset will never reflect the true lows or true highs. Prices are always increasing, there is no maximum or minimum value. Therefore, we may use the **Robust Scaler** approach which uses the median and InterQuartile Range (rather than the min-max) which makes it resistant to outliers.

```python
rs = RobustScaler()
def scaler_function (x):
    scaled_df = rs.fit_transform(x) 
    return scaled_df
```

#### 7.1.2 Function to split the historical transaction data

```python
def split_function(dataset):
    split_ratio = 0.8
    split_index = int(split_ratio * len(dataset))

    train = dataset[:split_index]
    test = dataset[split_index:]
    
    return train, test
```

#### 7.1.3 Function to build the Model using LSTM by adding different layers

We build the model containg 4 layers of LSTM network, all followed by a **Dropout layer**.  To reduce overfitting, the dropout layer just randomly takes a portion of the possible network connections. This value is the percentage of the considered network connections per epoch/batch. On the bottom, we also have a final **Dense layer**. 

```python
def create_model(inputshape):
    model = Sequential()

    #First Layer
    model.add(LSTM(units = 100, return_sequences=True, input_shape = inputshape)) #Layer of 100 units
    model.add(Dropout(0.2))

    #Second Layer
    model.add(LSTM(units = 100, return_sequences=True))
    model.add(Dropout(0.2))

    #Third Layer
    model.add(LSTM(units = 100, return_sequences=True))
    model.add(Dropout(0.2))

    #Fourth Layer
    model.add(LSTM(units = 100, return_sequences=False))
    model.add(Dropout(0.2))

    model.add(Dense(units=1)) #A dense layer with one output unit
    
    return model
```

   - **Initiate some empty placeholder lists to store results**

```python
mse = []
mae = []
model_name = []
```

### 7.2 LSTM - Single Feature
Extract the Returns column from the corresponding dataframe for training and testing based on this feature.

```python
"""
Create the training data set & the testing data set, based on one single feature (daily 
returns) that will be used as independent and dependent variable
"""
daily_returns = dfReturnsScore2[['Returns']].copy()
# dfReturns_sc = scaler_function(daily_returns) # NO need to scale the data at this point
ret_df = daily_returns.values
train_df, test_df = split_function(ret_df)
```

Since LSTM works on sequences using them as the base for prediction of a single value, a matrix must be created from the train data set available. We prepare input sequences, with **30 timesteps** of X_train data alongside the respective y_train labels. Thereafter, we reshape the data by adding another access for the batch size, since the input for an LSTM network is a 3D tensor. It must have a sequence length, the timesteps and the batch size. 

```python
"""
Create X_train and y_train data structures
"""

X_train = []
y_train = []
X_test = []

length_train = len(train_df)
length_test = len(test_df)

timestep = 30

for i in range(timestep, length_train): 
    X_train.append(train_df[i-timestep:i,0])
    y_train.append(train_df[i,0])

for i in range(timestep, length_test):
    X_test.append(test_df[i-timestep:i,0])

"""
Convert the X_train, y_train, X_test to numpy arrays
"""
   
X_train = np.array(X_train)
y_train = np.array(y_train)

print('X_train shape:', X_train.shape)
print('y_train shape:', y_train.shape)

X_test = np.array(X_test)
print('X_test shape:', X_test.shape)
print()

"""
Reshape the data
"""

X_train = np.reshape(X_train,(X_train.shape[0], X_train.shape[1], 1)) #adding the batch_size axis
print('X_train after reshape:', X_train.shape)
print('y_train shape:', y_train.shape)

X_test = np.reshape(X_test,(X_test.shape[0], X_test.shape[1],1))
print('X_test after reshape:', X_test.shape)
```

#### 7.2.1 Fit the Model

Train the data using 10 epochs (iterations) and batch_size of 32 using the `model.fit` function by parsing X_train and y_train data. Compile using an adam optimizer and the mean squared error as a loss function.

```python
input_shape = (X_train.shape[1],1)
model_single = create_model(input_shape) #We use the model function and the corresponding parameter input_shape 
model_single.compile(optimizer = 'adam', loss = 'mean_squared_error')
model_single.fit(X_train, y_train, epochs=10, batch_size=32, verbose=2)

"""
'loss' refers to the loss value over the training data after each epoch. 
This is what the optimization process is trying to minimize with the training so, the lower, the better.
"""
```

- Predict the model output using the `model.predict` function by passing the X_test data

```python
y_pred = model_single.predict(X_test)

#Skip this step: 
#Set the proper format of the results using `inverse_transform` function to plot the values with the actual and predicted data
    # y_pred_normal = rs.inverse_transform(y_pred)
    #test_returns = rs.inverse_transform(test_df[30:])

print('predicted data shape (single feature):', y_pred.shape)
print('test data shape:', test_df[timestep:].shape)

#add results to the list 
mse.append(mean_squared_error(test_df[timestep:], y_pred))
mae.append(mean_absolute_error(test_df[timestep:], y_pred))
model_name.append('LSTM - Stock Returns')
```

## 8. Setting up & testing the Trading strategy
The trading strategy to be used is called the **Simple Moving Average** crossover. 

In this section, we work with Rolling means, which help the algorithm evaluate the current price or stock return against the average price (return) a week, bi-week, month, term, year, or any other span. For example, **if the price now is higher than the annual price, it could mean that the stock is on an upward trend**. 

To calculate the rolling averages, we'll use the pandas rolling method to find the rolling mean of the Returns column for different time horizons. For example, we can create a short-term average by using a 7 day simple moving average, as well as a long-term average by using a 14 day simple moving average. 


```python
#Having a glance. Check df from bullet point 7.2
#daily_returns
```

### Using final compound VADER scores with threshold to generate trade calls

In general and mostly, the compound score is used as a threshold value for the analysis of the text data. We can even leverage the flexibility of changing the compound score (threshold value) and label the text data. This merely depends on business requirements and domain knowledge. 

Below is the standard scoring metric followed by most of the analyzers.

    -Positive sentiment: compound score >= 0.05
    -Neutral sentiment: (compound score > -0.05) and (compound score < 0.05)
    -Negative sentiment: compound score <= -0.05
    
We'll use a threshold value for making trade calls in our model. For days where there is a volatile behavior of markets, we may use 0.20. Use df from bullet point 6.2.4

```python
# VADER trade calls (with threshold)
vader_Buy=[]
vader_Sell=[]
threshold = 0.05

for i in range(len(dfReturnsScore2.tail(sample_size))):
    if dfReturnsScore2['Score'].values[i] > threshold:
        print("Trade Call for {row} is Buy.".format(row=dfReturnsScore2.index[i].date()))
        vader_Buy.append(dfReturnsScore2.index[i].date())
    elif dfReturnsScore2['Score'].values[i] < - (threshold):
        print("Trade Call for {row} is Sell.".format(row=dfReturnsScore2.index[i].date()))
        vader_Sell.append(dfReturnsScore2.index[i].date())
```

Merging Trade Signals with SMA at higher priority and VADER for Refining

```python
vader_buy = []
for i in range(len(signal)):
    if signal.index[i].date() in vader_Buy:
        vader_buy.append(i)
        
vader_sell = []
for i in range(len(signal)):
    if signal.index[i].date() in vader_Sell:
        vader_sell.append(i)
```

```python
#prioritising SMA signals
final_buy = list(set(Trade_Buy + vader_buy) - set(Trade_Sell)) 
final_sell = list(set(Trade_Sell + vader_sell) - set(Trade_Buy))

plt.figure(figsize=(20, 10),dpi=80)
plt.plot(signal.index, signal['SMA'+str(short_term)], color='dodgerblue') #, color='purple'
plt.plot(signal.index, signal['SMA'+str(long_term)], color='darkorange') #, color='orange'
plt.plot(signal.index, signal['Returns'],'-^', markevery=final_buy, ms=15, color='green') 
plt.plot(signal.index, signal['Returns'],'-v', markevery=final_sell, ms=15, color='red') 
plt.plot(signal.index, signal['Returns'], color='powderblue')
plt.xlabel('Date',fontsize=14)
plt.ylabel('Returns in Dollars', fontsize = 14)
plt.xticks(rotation='60',fontsize=12)
plt.yticks(fontsize=12)
plt.title('Trade Calls - MERGED', fontsize = 16)
plt.legend(['SMA'+str(short_term),'SMA'+str(long_term),'Buy','Sell','Returns'])
plt.grid()
plt.show()
```

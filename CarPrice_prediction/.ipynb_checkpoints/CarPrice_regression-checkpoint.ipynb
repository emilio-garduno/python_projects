## 4. Feature Selection

    a. Assign the price column of the dataframe to a new variable called target, i.e., target = df.price.
    In addition to that, create another variable to save the features except the price. 


```python
df1 = df.copy()
```

```python
#Divide the df into target and features
target = df1.price #dependent variable
features = df1.drop('price', axis=1)
```

    b. Correlation Based Selection (manual or threshold)

```python
cor = df1.corr()
cor_target = abs(cor['price']) #absolute value of the correlation with price. We ignore the diagonal values because that is correlation of variable with itself.
relevant_features = cor_target[cor_target>0.12] #Threshold, correlation > 0.12
relevant_features = relevant_features.sort_values(ascending = False) #sort the output. 
relevant_features
```

```python
df_corr = features.copy()
```

```python
df_corr.drop(columns=['body_coupe','body_hatchback','body_minivan','body_van','body_wagon','trans_manual','make_acura',
                'make_audi','make_bmw','make_cadillac','make_chevrolet','make_chrysler','make_dodge','make_ford',
               'make_gmc','make_honda','make_hyundai','make_jeep','make_kia','make_lexus','make_lincoln',
               'make_mazda','make_mercedes_benz','make_nissan','make_toyota','make_volkswagen',
               'color_black','color_blue','color_gray','color_red','color_silver','color_white',
               'inter_beige','inter_tan'], inplace = True)
```

```python
print(df_corr.shape)
```

```python
#Correlation between independent variables?
#sns.pairplot(df_corr, height=2.5)
#plt.tight_layout()
```

    c. Variance Threshold Selection

```python
#We remove features that do not vary a lot, because they may not affect the model
from sklearn.feature_selection import VarianceThreshold

thresholds = np.arange(0.1,0.2)
for t in thresholds:
#VarianceThreshold object
    vt = VarianceThreshold(threshold=t)
    vt.fit(features) #fit finds the features with variance between 0.1 and 0.2 
pd.DataFrame({'Variance':vt.variances_, 'Status':vt.get_support()}, index = features.columns).T
```

```python
df_vt = features.iloc[:,vt.get_support()]
print(df_vt.shape)
df_vt.head(5)
```

    d. Select K-Best method

```python
from sklearn.feature_selection import SelectKBest #SelectKBest find the best k feature
from sklearn.feature_selection import f_regression #We use 'f_regression' for regression algorithm
from sklearn.model_selection import train_test_split

kfeat = np.arange(8,14) #Select between 8 and 14 
for i in kfeat:
    kbest = SelectKBest(score_func=f_regression, k=i)
#score_func: here we can use f_regression or f_classification. k represents the number of features between 8 and 14. 
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.25, random_state=42)

kbest.fit_transform(X_train, y_train) #fit_transform is for training
df_selKBest = features.iloc[:,kbest.get_support()] #for feature selection we use get_support
print(df_selKBest.shape)
df_selKBest.head()
```

## 6. Linear Regression Model with Lasso

```python
X_trainL, X_testL, Y_trainL, Y_testL = train_test_split(features, target, test_size = 0.25, random_state = 42)
                                                    #Here, the first value refers to X, the second refers to Y

# X: features
# Y: target
```

```python
#Play around with the values of alpha, the number of alphas being tested, and the number iteration of the Lasso 
alphasL = np.linspace(1, 0.001, 10)
alphasL
```

```python
rmse_lassoList = []
r2_lassoList = []
model_lassoList = []

for a in alphasL:
    lasso = Lasso(alpha=a, max_iter=500)
    lasso.fit(X_trainL, Y_trainL)
    predL = lasso.predict(X_testL)
    
    r2_lassoList.append(lasso.score(X_trainL, Y_trainL))
    rmseL = mean_squared_error(Y_testL, predL)**0.5;
    rmse_lassoList.append(rmseL)
    model_lassoList.append(lasso)
    #print("Alpha", a, "RMSE", rmseL)
    
lasso_df = pd.DataFrame(zip(alphasL, rmse_lassoList, r2_lassoList), columns=['Alpha','RMSE',"R2"])
lasso_df
```

```python
#Set a variable to display a frame with alpha with the min RMSE and the max R2
best_alpha_lasso = lasso_df.sort_values(by=['RMSE', 'R2'], ascending=[True, False]).head(1) 
best_alpha_lasso.head()
```

```python
a_lasso = best_alpha_lasso.iat[0,0] 
rmse_lasso = best_alpha_lasso.iat[0,1] #we use iat to get the value at specified row/column pair. Row [0] Column [1]
r2_lasso = best_alpha_lasso.iat[0,2]
#r2_lasso
#rmse_lasso
```

```python
feat_sel.append("Lasso, alpha = " + str(a_lasso.round(decimals = 3)))
transf_list.append("None")
scal_list.append("None")
r2_list.append(r2_lasso)
rmse_list.append(rmse_lasso)
```

```python
X_trainR, X_testR, Y_trainR, Y_testR = train_test_split(features, target, test_size = 0.25, random_state = 42)
                                                    #Here, the first value refers to X, the second refers to Y

# X: features
# Y: target
```

```python
#Play around with the values of alpha, the number of alphas being tested, and the number iteration of the Ridge
alphasR = np.linspace(1, -1, 10)
alphasR
```

```python
rmse_ridgeList = []
r2_ridgeList = []
model_ridgeList = []

for a in alphasR:
    ridge = Ridge(alpha=a, max_iter=1000) #max iterations 1000
    ridge.fit(X_trainR, Y_trainR)
    predR = ridge.predict(X_testR) #predict
    
    r2_ridgeList.append(ridge.score(X_trainR,Y_trainR))
    rmseR = mean_squared_error(Y_testR, predR)**0.5; #**means exponentiation
    rmse_ridgeList.append(rmseR)
    model_ridgeList.append(ridge)
    #print("Alpha",a,"RMSE",rmse)
    
ridge_result = np.vstack((alphasR, rmse_ridgeList, r2_ridgeList)).T
ridge_df = pd.DataFrame(ridge_result, columns=['Alpha', 'RMSE', "R2"])
ridge_df
```
